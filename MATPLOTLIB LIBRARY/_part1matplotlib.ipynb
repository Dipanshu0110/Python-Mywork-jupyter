MATPLOTLIB LIBRARY


```python
import matplotlib.pyplot as plt
```

LINEAR GRAPH


```python
x=[1,2,3,4]

y=[5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
#change color
x=[1,2,3,4]
y=[5,6,7,8]
c='r'
plt.plot(x,y,c)
plt.show()
```

```python
x=[5,2,9,4,7]
y=[7,4,2,10,12]
plt.plot(x,y,marker='o')
plt.show()
```

```python
x=[5,2,9,4,7]
y=[7,4,2,10,12]
plt.plot(x,y)
plt.plot(x)
plt.plot(y,marker='1')
plt.show()
```

BAR GARPH

```python
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','g','y','b']
plt.bar(x,y, color = c)
plt.show()
```

```python
import numpy as np
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.barh(x,y,color='hotpink')
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y,color='hotpink',width=0.1)
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.barh(x,y,color='hotpink',height=0.1)
plt.show()
```

SCATTER PLOT

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y,color='black')
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y,color='black', s=1150)
plt.show()
```
