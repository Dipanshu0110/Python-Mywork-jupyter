BARPLOT IN SEABORN LIBRARY

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1=sns.load_dataset('penguins')
```

```python
df1
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1)
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex')
```

```python
order=['Dream','Torgersen','Biscoe']
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,order=order,hue='sex')
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,order=order,hue='sex',palette='gist_earth')
```

SATURATION

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,order=order,hue='sex' ,palette='gist_earth',saturation=0.1)
```
